```python
#Import Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Create DataFrame
data = [100, 150, 50, 100, 130, 120, 100, 85, 70, 150,
        150, 120, 50, 100, 100, 140, 90, 150, 50, 90,
        120, 100, 110, 75, 65]
df = pd.DataFrame({'TransactionAmount': data})
print("--- DataFrame 'TransactionAmount' ---")
print(df)
print(f"\nJumlah data: {len(df)}")
print("-" * 30)
```

```python
#Central Tendency
# --- Langkah 2: Analisis Measure of Central Tendency ---
print("\n--- Analisis Measure of Central Tendency ---")
# Menghitung Mean (Rata-rata)
mean_value = df['TransactionAmount'].mean()
print(f"Mean (Rata-rata): ${mean_value:.2f}")
# Menghitung Median (Nilai Tengah)
median_value = df['TransactionAmount'].median()
print(f"Median (Nilai Tengah): ${median_value:.2f}")
# Menghitung Modus (Nilai yang Paling Sering Muncul)
# mode() mengembalikan Series, karena bisa ada lebih dari satu modus
mode_values = df['TransactionAmount'].mode()
# Mengubah Series ke list string untuk dicetak
mode_str = ', $'.join(map(str, mode_values.tolist()))
print(f"Modus (Nilai Paling Sering Muncul): ${mode_str}")
print("-" * 30)
```

```python
#Tulis analisismu dari Central Tendency di sini

"""
Analisis Mean, Median, dan Modus
Anda telah memberikan nilai-nilai berikut:

Mean (Rata-rata): $102.60** * **Median (Nilai Tengah): $$102.60** * **Median (Nilai Tengah): $100.00
Modus (Nilai yang Paling Sering Muncul): $100
Nilai-nilai ini memberikan wawasan tentang jumlah transaksi tipikal dalam dataset. Berikut uraiannya:

Mean: Rata-rata jumlah transaksi adalah $102.60. Ini adalah rata-rata dari semua nilai transaksi. Nilainya sedikit lebih tinggi dari median, 
      menunjukkan bahwa mungkin ada beberapa jumlah transaksi yang lebih tinggi yang menarik rata-rata ke atas.

Median: Median jumlah transaksi adalah $100.00. Ini adalah nilai tengah ketika semua jumlah transaksi diurutkan. Karena lebih rendah dari mean, 
      ini menunjukkan bahwa distribusi jumlah transaksi mungkin sedikit miring ke kanan (positively skewed), yang berarti ada beberapa jumlah 
      transaksi yang lebih besar yang memengaruhi mean lebih dari median.

Modus: Modus jumlah transaksi adalah $100. Ini adalah nilai yang paling sering muncul dalam dataset. Dalam kasus ini, modus sama dengan median, 
      yang semakin memperkuat indikasi bahwa $$100. Ini adalah nilai yang paling sering muncul dalam dataset. Dalam kasus ini, modus sama dengan median, 
      yang semakin memperkuat indikasi bahwa $100 adalah nilai yang representatif untuk jumlah transaksi tipikal.

Interpretasi Keseluruhan:

Mean, median, dan modus yang relatif berdekatan satu sama lain menunjukkan bahwa data terdistribusi agak simetris di sekitar nilai pusat sekitar $100 hingga $$100 hingga $102.
Namun, sedikit perbedaan antara mean dan median mengisyaratkan kemungkinan adanya sedikit kemiringan positif (positive skew), yang berarti ada beberapa transaksi
yang lebih besar yang sedikit memengaruhi mean.

Fakta bahwa modus sama dengan median ($100) semakin memperkuat bahwa nilai tersebut merupakan representasi yang baik dari pusat data dan jumlah transaksi yang paling sering terjadi.

Kesimpulan:

Secara umum, dapat disimpulkan bahwa jumlah transaksi tipikal dalam dataset berada di sekitar $100. Meskipun ada beberapa transaksi yang lebih besar, mereka tidak secara signifikan
mengubah gambaran keseluruhan distribusi data. $$100. Meskipun ada beberapa transaksi yang lebih besar, mereka tidak secara signifikan mengubah gambaran keseluruhan distribusi data. 
$100 dapat dianggap sebagai nilai yang representatif untuk jumlah transaksi yang paling sering terjadi dan nilai tengah dari seluruh data transaksi.
"""
```

```python
#Variability

# Calculating Variance (var)
variance = df['TransactionAmount'].var()
print(f"Variance: {variance:.2f}")

# Calculating Standard Deviation (std)
std_dev = df['TransactionAmount'].std()
print(f"Standard Deviation: {std_dev:.2f}")

# Getting Minimum and Maximum Values (min, max)
min_value = df['TransactionAmount'].min()
max_value = df['TransactionAmount'].max()
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")

# Calculating Quantiles (0.25, 0.5, 0.75)
quantile_25 = df['TransactionAmount'].quantile(0.25)
quantile_50 = df['TransactionAmount'].quantile(0.5)
quantile_75 = df['TransactionAmount'].quantile(0.75)
print(f"25th Quantile: {quantile_25}")
print(f"50th Quantile: {quantile_50}")
print(f"75th Quantile: {quantile_75}")
```

```python

#Tulis analisismu dari Variability di sini

"""
Analisis Variability

Berikut adalah nilai-nilai variability yang telah dihitung:

Variance: 1012.75
Variance menunjukkan sebaran data di sekitar rata-rata. Nilai variance yang tinggi (1012.75) mengindikasikan bahwa data transaksi
memiliki sebaran yang cukup luas di sekitar nilai rata-rata.
Standard Deviation: 31.82
Standard deviation adalah akar kuadrat dari variance dan lebih mudah diinterpretasikan. Nilai 31.82 menunjukkan bahwa, rata-rata, 
setiap data transaksi menyimpang sekitar 31.82 unit dari nilai rata-rata.

Minimum Value: 50
Nilai transaksi terkecil dalam dataset adalah 50.

Maximum Value: 150
Nilai transaksi terbesar dalam dataset adalah 150.

25th Quantile (Q1): 85.0
25% dari data transaksi memiliki nilai kurang dari atau sama dengan 85.0.

50th Quantile (Q2/Median): 100.0
50% dari data transaksi memiliki nilai kurang dari atau sama dengan 100.0. Ini juga merupakan nilai median.

75th Quantile (Q3): 120.0
75% dari data transaksi memiliki nilai kurang dari atau sama dengan 120.0.

Interpretasi Keseluruhan
Dari nilai-nilai ini, kita dapat menyimpulkan bahwa:

Data Transaksi Cukup Tersebar: Variance dan standard deviation yang relatif tinggi menunjukkan bahwa data transaksi cukup bervariasi dan tidak terpusat di sekitar nilai rata-rata.
Rentang Data: Data transaksi memiliki rentang antara 50 hingga 150, menunjukkan perbedaan yang signifikan antara transaksi terkecil dan terbesar.
Sebaran Data: Kuartil memberikan gambaran tentang sebaran data. 50% data transaksi berada di antara 85.0 dan 120.0 (Q1 dan Q3).
"""
```

```python
#Define H0 and H1
# H0: mu <= 100 penjualan rata-rata yaitu $100 per transaksi
# H1: mu != / > 100 penjualan rata-rata yaitu meningkat
mu0 = 100 # Hypothesized mean
```

```python
#Define alpha = 0.05
alpha = 0.05
```

```python
#Calculate t statistics and P-value
#Perform one sample t-test
t_statistic, p_value = stats.ttest_1samp(df['TransactionAmount'], mu0)
print(f"Rata-rata Populasi (H0): ${mu0}")
print(f"Rata-rata Sampel: ${mean_value:.2f}")
print(f"Tingkat Signifikansi (alpha): {alpha}")
```

```python
#statistic= , pvalue=

print(f"T-Statistic: {t_statistic:.4f}")

print(f"P-value (one-tailed, greater): {p_value:.4f}")
```

```python
#Conclude with alpha and p-value

#Conclusion: 
if p_value < alpha:
    print(f"\nHasil: Menolak H0 (p-value < alpha)")
    print("Kesimpulan Sementara: Terdapat cukup bukti statistik pada tingkat signifikansi 5% untuk"
           "menyatakan bahwa rata-rata jumlah transaksi populasi tidak sama dengan 100 meningkat setelah training.")
else:
    print(f"\nHasil: Gagal Menolak H0 (p-value >= alpha)")
    print("Kesimpulan Sementara: Tidak terdapat cukup bukti statistik pada tingkat signifikansi 5% untuk"
          "menyatakan bahwa rata-rata jumlah transaksi populasi berbeda secara signifikan dari 100 transaksi meningkat setelah training.")
```

```python
#Conclude with t statistics and critical region
#Find the critical region
n = len(df['TransactionAmount'])
dof = n - 1
t_critical_lower = stats.t.ppf(alpha/2, dof)
t_critical_upper = stats.t.ppf(1 - alpha/2, dof)

print(f"T-Statistic: {t_statistic:.4f}")

print(f"Nilai Kritis T (untuk α={alpha}, two-tailed): {t_critical_lower:.4f} dan {t_critical_upper:.4f}")

# Hasil Daerah Kritis dan Kesimpulan:
print("\n--- Hasil Daerah Kritis ---")
if t_statistic < t_critical_lower or t_statistic > t_critical_upper:
    print(f"Karena statistik uji T ({t_statistic:.4f}) berada di dalam daerah kritis (t < {t_critical_lower:.4f} atau t > {t_critical_upper:.4f}), kita menolak Hipotesis Nol (H0)")
    print("Kesimpulan :Terdapat cukup bukti statistik untuk mendukung Hipotesis Alternatif (H1) bahwa rata-rata jumlah transaksi populasi tidak sama dengan 100.")
else:
    print(f"Karena statistik uji T ({t_statistic:.4f}) tidak berada di dalam daerah kritis (antara {t_critical_lower:.4f} dan {t_critical_upper:.4f}), kita gagal menolak Hipotesis Nol (H0).")
    print("Kesimpulan : Tidak terdapat cukup bukti statistik untuk mendukung Hipotesis Alternatif (H1) bahwa rata-rata jumlah transaksi populasi tidak sama dengan 100.")
```
